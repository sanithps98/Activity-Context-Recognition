### **Selecting necessary features from Attributes**

```python
#Printing out the attributes
col = list(df.columns)
required_features = []
other_features = []
for i in col:
    if ((i != "lux") and (i != "soundLevel")):
        required_features.append(i)
    else:
        other_features.append(i)

print('Required Features :',*required_features)
print('Other Features :',*other_features)
```

*   Since light and sound sensor are very less relevant for determining the activity of the subject, it is better to not consider the two attributes: **lux** and **soundLevel** for activity context recognition.


### **Distribution and Visualisation of Target Variable - Activity**

```python
# Plot the value_counts with a bar graph
ax = df.activity.value_counts().plot(kind="bar")
plt.title('User Activity');

ax.set_xlabel("ACTIVITY")
ax.set_ylabel("COUNT")
for rect in ax.patches:
    y_value = rect.get_height() 
    x_value = rect.get_x() + rect.get_width() / 2
    space = 1
    label = "{:.0f}".format(y_value)
    ax.annotate(label, (x_value, y_value), xytext=(0, space), textcoords="offset points", ha='center', va='bottom')
plt.show()
```

##### **2) Data Standardisation**

```python
X = df2
display(X.head())
```

```python
y = df.iloc[:,-1]
y.unique()
```

```python
#Standardising the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)
X = pd.DataFrame(rescaledX, columns=X.columns)
X.head()
```

```python
display(X.shape)
```

##### **3) SMOTE**

```python
#Applying SMOTE Technique
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state = 14)
X_smote, y_smote = smote.fit_resample(X, y)
```

```python
# Plot the value_counts with a bar graph
ax = y_smote.value_counts().plot(kind="bar")
plt.title('User Activity');

ax.set_xlabel("ACTIVITY")
ax.set_ylabel("COUNT")
for rect in ax.patches:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
plt.show()
```

```python
y_smote = y_smote.astype('category').cat.codes
y_smote.unique()
```

```python
datasetDf = X_smote
datasetDf['category'] = y_smote
datasetDf.head()
```
