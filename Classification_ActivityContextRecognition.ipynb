### **Functions**

```python
modelPerformance_table_dataFrame = pd.DataFrame(data=[['Model', 100, 1, 1, 1]], 
                            columns=['Model', 'Accuracy', 'F1_Score', 'Precision', 'Recall'])
```

```python
def modelPerformace(y_values, predicted_values, model_name, reset_table):
  print("Test Result:\n----------------------------")
  # Accuracy Score
  from sklearn import metrics
  accuracy_score = metrics.accuracy_score(y_values,predicted_values)
  print('Accuracy Score:', accuracy_score)

  # F1 Score
  from sklearn.metrics import  f1_score
  f_score = f1_score(y_values, predicted_values, average='micro')
  print("F1 Score:", f_score)

  # Precision SetValue
  from sklearn.metrics import precision_score
  precision_score_value = precision_score(y_values, predicted_values, average='micro')
  print("Precision Value:",precision_score_value)

  # Recall ValueSet
  from sklearn.metrics import recall_score
  recall_score_value = recall_score(y_values, predicted_values, average='micro')
  print("Recall Value:",recall_score_value)

  if(reset_table):
    global modelPerformance_table_dataFrame
    modelPerformance_table_dataFrame = pd.DataFrame(data=[[model_name, accuracy_score, f_score, precision_score_value, recall_score_value]], 
                            columns=['Model', 'Accuracy', 'F1_Score', 'Precision', 'Recall'])
  else:
    results = pd.DataFrame(data=[[model_name, accuracy_score, f_score, precision_score_value, recall_score_value]], 
                          columns=['Model', 'Accuracy', 'F1_Score', 'Precision', 'Recall'])
    modelPerformance_table_dataFrame = modelPerformance_table_dataFrame.append(results, ignore_index=True)

  print()
  # Confusion Matrix
  from sklearn.metrics import confusion_matrix
  %matplotlib inline
  cm = confusion_matrix(y_values, predicted_values)
  f,ax = plt.subplots(figsize=(5, 5))
  sns.heatmap(cm, annot=True, linewidths=0.5,linecolor="red", fmt= '.0f',ax=ax)
  plt.title('Confusion Matrix ' + model_name)
  plt.show()
```

```python
def DecisionTreeClassifierModel(X, y, test):
  from sklearn.tree import DecisionTreeClassifier
  decisionTreeClassifier = DecisionTreeClassifier(max_depth = 20)
  decisionTreeClassifier.fit(X, y)
  return decisionTreeClassifier.predict(test)
```

```python
def randomForestModel(X, y, test):
  from sklearn.ensemble import RandomForestClassifier
  classifierRandomForest = RandomForestClassifier(n_estimators = 10, random_state = 0)
  classifierRandomForest.fit(X, y)
  return classifierRandomForest.predict(test)
```

```python
def MLPModel(X, y, test):
  from sklearn.neural_network import MLPClassifier
  mlp = MLPClassifier(hidden_layer_sizes=(20))
  mlp.fit(X, np.array(y))
  return mlp.predict(test)
```

**Splitting Standardized Dataset into Training and Testing Set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

### **Classification Models**

```python
y_pred_DecisionTreeClassifier = DecisionTreeClassifierModel(X_train, y_train, X_test)
```

```python
modelPerformace(y_test, y_pred_DecisionTreeClassifier, 'Decision Tree Classifier', True)
```

```python
y_pred_RandomForest = randomForestModel(X_train, y_train, X_test)
```

```python
modelPerformace(y_test, y_pred_RandomForest, 'Random Forest Classifier', False)
```

```python
y_pred_Mlp = MLPModel(X_train, y_train, X_test)
```

```python
modelPerformace(y_test, y_pred_Mlp, 'Multi-Layer Perceptron Classifier', False)
```
